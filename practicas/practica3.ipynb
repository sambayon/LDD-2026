```python
import seaborn.objects as so
import pandas as pd
gapminder = pd.read_csv("gapminder.csv")

datosArgentina = gapminder[gapminder['country'] == 'Argentina'].reset_index()
datosArgentina
```

```python
(
    so.Plot(
        data = datosArgentina,
        x = "year",
        y = 'lifeExp'
    )
    .add(
        so.Line()
    )
    .label(
        title= 'Expectativa de vida de Argentina por año',
        x = 'Año',
        y = 'Edad'
    )
)
```

```python
datosAmericas = gapminder[gapminder["continent"] == "Americas"]

(
    so.Plot(
        data=datosAmericas, 
        x="year", 
        y="lifeExp"
        )
    .add(
        so.Lines(color="#bbca"), 
        group="country"
        )
    .add(
        so.Line(), 
        so.PolyFit(1)
        )
)
    
```

```python
(
    so.Plot(
        data = gapminder,
        x = 'year',
        y = 'lifeExp',
        group = 'continent'
    )
    .add(
        so.Lines(color = '#bbca'),
        group = 'country'
    )
    .add(
        so.Line(),
        so.PolyFit(1)
    )
    .facet(col = "continent", wrap = 3)
)
```

Estas estudiando la relacion entre altura y peso de las personas.  
Tenes un data-set que tiene como variables la edad, sexo y peso de cada persona.  
que grafico harıas para cada una?  ¿y si queres visualizar la relacion entre peso y altura?  

Peso - Altura - Edad = **Histograma**

Sexo = **Barras**

Relacion peso altura = **Scatter Plot**

```python
import numpy as np
# Fijamos una cantidad de personas para el DataFrame chiquito (ej. 50 personas)
n_personas = 50

# 1. Altura random entre 160 cm y 190 cm (valores continuos)
altura = np.random.uniform(160, 190, n_personas)

# 2. Edad entre 18 y 70 años (valores enteros)
edad = np.random.randint(18, 71, n_personas)

# 3. Peso entre 60 kg y 100 kg (valores continuos)
peso = np.random.uniform(60, 100, n_personas)

# 4. Sexo con variabilidad random de F entre 30% y 60%
# Primero elegimos una proporción fija para este DataFrame específico
prob_femenino = np.random.uniform(0.30, 0.60)
prob_masculino = 1 - prob_femenino

sexo = np.random.choice(
    ["F", "M"], size=n_personas, p=[prob_femenino, prob_masculino]
)

# Creamos el DataFrame
df = pd.DataFrame(
    {"edad": edad, "sexo": sexo, "peso": np.round(peso, 1), "altura": np.round(altura, 1)}
)

# Ver los primeros datos y la proporción real que tocó de cada sexo
print(df.head())
print("\nProporción real en este DataFrame:")
print(df["sexo"].value_counts(normalize=True))
```

```python

(
    so.Plot(data=df, x="peso")
    .add(so.Bars(), so.Hist())
    .label(title="Distribución del Peso", x="Peso (kg)", y="Cantidad de personas")
)
```

```python
(
    so.Plot(
        data=df, x="sexo"
        )
    .add(
        so.Bar(), so.Hist()
        ) 
)
```

```python
(
    so.Plot(data=df, x="altura", y="peso", color = 'sexo')
    .add(so.Dot())
    .label(title="Relación entre Altura y Peso", x="Altura (cm)", y="Peso (kg)")
)
```
