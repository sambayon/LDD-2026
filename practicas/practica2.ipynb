# Archivo de Datos

```python
df = pd.read_csv("casos_coronavirus.csv")
df

df["confirmados_Nuevos"].plot()
```

```python
import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt
```

```python
df["confirmados_Nuevos"].plot()

plt.title("Casos de Coronavirus por Día")
plt.xlabel("Días transcurridos")
plt.ylabel("Cantidad de casos nuevos")
plt.tight_layout()
plt.savefig("casos_por_dia.png")
```

```python
df['casos_Acumulados'] = df['confirmados_Nuevos'].cumsum()

df["casos_Acumulados"].plot()
plt.title("Casos Acumulados de Coronavirus")
plt.xlabel("Días transcurridos")
plt.ylabel("Cantidad de casos acumulados")
plt.tight_layout()
plt.savefig("casos_acumulados.png")
```

# Escala logaritmica 

```python
df['y'] = np.log(df['casos_Acumulados'])

df["dias_transcurridos"] = df.index

df.plot(x="dias_transcurridos", y="y", legend=False)
plt.title("Logaritmo de Casos Acumulados vs. Días Transcurridos")
plt.xlabel("Días transcurridos")
plt.ylabel("$y = \log(\text{Casos Acumulados})$")
plt.tight_layout()
plt.savefig("log_casos_acumulados.png")
```
